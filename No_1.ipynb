```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
from google.colab import drive
drive.mount('/content/drive')
f = open('./drive/MyDrive/Colab Notebooks/UTS/teks pidato jokowi 2022.txt', 'r')
isi_berita = f.read()
print(isi_berita)
```

```python
f2 = open('./drive/MyDrive/Colab Notebooks/UTS/tes pidato presiden jokowi 2023.txt', 'r')
isi_berita2 = f2.read()
print(isi_berita2)
```

```python
word_tokens = word_tokenize(isi_berita)
print(word_tokens)
```

```python
word_tokens2 = word_tokenize(isi_berita2)
print(word_tokens2)
```

```python
word_tokens=[word.lower() for word in word_tokens if word.isalpha()]
word_tokens2=[word.lower() for word in word_tokens2 if word.isalpha()]
```

```python
stop_words = set(stopwords.words('indonesian'))
word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
print(word_tokens_no_stopwords)
```

```python
word_tokens_no_stopwords_2 = [w for w in word_tokens2 if not w in stop_words]
print(word_tokens_no_stopwords_2)
```

```python
freq_kata = nltk.FreqDist(word_tokens_no_stopwords)
print(freq_kata.most_common(20))
```

```python
freq_kata2 = nltk.FreqDist(word_tokens_no_stopwords_2)
print(freq_kata2.most_common(20))
```

```python

import matplotlib.pyplot as plt
freq_kata.plot(10)
plt.show()
freq_kata2.plot(10)
plt.show()
```
